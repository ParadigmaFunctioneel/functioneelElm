<h1>Algemene Uitwerking Vraag1</h1>

```elm
import Html exposing(..)
main = text ("Afstand = " ++ (toString afstand) ++ " m")

c = 3.0*(10^8)
milli g = g * 10^(-3)

s_signaal t = c * t
s_object t = 0.5 * (s_signaal t)

afstand = s_object (milli 0.26)
-- compile-code
```

<h1>Algemene Uitwerking Vraag3</h1>

```elm
import Html exposing(..)

main = text ("Afmeting = " ++ (toString afmeting) ++ " m")

c = 3.0 * 10^8
giga g = g * 10^9
procent p g = p/100 * g

gl s f = s / f

afmeting = procent 10 (gl c (giga 9.38))
-- compile-code
```

<h1>Algemene Uitwerking Vraag4</h1>

```elm
import Html exposing(..)

main = text ("Oppervlakte = " ++ (toString oppervlakte))

kilo g = g * 10^3

opp a r1 r2 = a * (r2^4 / r1^4)

oppervlakte = opp 6.0 (kilo 30) (kilo 45)
-- compile-code
```
